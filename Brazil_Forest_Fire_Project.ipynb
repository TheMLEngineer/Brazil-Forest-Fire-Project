```
#Date column Gives same O/P as Year column so can be made index as index have zero effect on ML algorithm
```

```
df.head()
```

```
df.set_index('date' , drop = True , inplace= True)
```

```
df.head()
```

```
df.info()
```

```
for i in df:
  print(df[i].value_counts())
```

```
#Except numbers other can be classified as a categorical featires
```

```
df['year'] = df['year'].astype('category')
```

```
df['state'] = df['state'].astype('category')
```

```
df['month'] = df['month'].astype('category')
```

```
df.info()
```

```
model_df = pd.get_dummies(df)
```

```
model_df.head()
```

```
#Starting to do Machine Learning Part
```

```
model_df.info()
```

```
model_df.describe()
```

```
#Importing Necessary Libs
```

```
from sklearn.linear_model import LinearRegression
```

```
from sklearn.metrics import accuracy_score
```

```
# Splitting data in Correct format
```

```
model_df.head()
```

```
X = model_df.drop('number' , axis = 1)
X.head()
```

```
y = model_df['number']
```

```
y.tail()
```

```
from sklearn.model_selection import train_test_split
```

```
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state = 42)
```

```
X_train.shape
```

```
y_train.shape
```

```
X_test.shape
```

```
y_test.shape
```

```
#Implementing Linear Regression
```

```
linear_regression_obj = LinearRegression()
```

```
linear_regression_obj.fit(X_train , y_train)
```

```
y_pred = linear_regression_obj.predict(X_test)
```

```
y_pred
```

```
linear_regression_obj.coef_
```

```
linear_regression_obj.score(X_test , y_test)
```

```
print("Mean squared error: %.2f" % np.mean((linear_regression_obj.predict(X_test) - y_test) ** 2))
```

```
from sklearn import metrics
metrics.explained_variance_score(y_test, y_pred)
```
